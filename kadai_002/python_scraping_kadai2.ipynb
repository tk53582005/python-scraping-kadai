```python
import requests
import pprint
from getpass import getpass

print("ライブラリのインポート完了")
```

```python
api_key = getpass('Google Maps APIキーを入力: ')
```

```python
station_name = input('駅名を入力: ')
```

```python
geocoding_url = 'https://maps.googleapis.com/maps/api/geocode/json'
geocoding_params = {
    'address': station_name,
    'key': api_key
}

print(f"\n{station_name}の位置情報を取得中")
geocoding_response = requests.get(geocoding_url, params=geocoding_params)

if geocoding_response.status_code == 200:
    geocoding_data = geocoding_response.json()

    print("\nGeocoding API レスポンス")
    pprint.pprint(geocoding_data)

    if geocoding_data['status'] == 'OK' and geocoding_data['results']:
        location = geocoding_data['results'][0]['geometry']['location']
        latitude = location['lat']
        longitude = location['lng']

        print(f"\n宝塚駅の位置情報:")
        print(f"緯度: {latitude}")
        print(f"経度: {longitude}")
    else:
        print("位置情報の取得失敗")
else:
    print(f"リクエストエラー: {geocoding_response.status_code}")
```

```python
places_url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'
places_params = {
    'location': f'{latitude},{longitude}',
    'radius': 500,
    'type': 'restaurant',
    'language': 'ja',
    'key': api_key
}

print(f"\n{station_name}駅周辺500m以内のレストランを検索中")
places_response = requests.get(places_url, params=places_params)

if places_response.status_code == 200:
    places_data = places_response.json()

    print("\nPlaces API レスポンス")
    pprint.pprint(places_data)
else:
    print(f"Places APIリクエストエラー: {places_response.status_code}")
```

```python
if places_data['status'] == 'OK':
    restaurants = places_data['results']

    print(f"\n" + "=" * 60)
    print(f"{station_name}駅周辺のレストラン情報")
    print(f"検索範囲: 半径500m以内")
    print(f"見つかったレストラン数: {len(restaurants)}件")
    print("=" * 60)

    for i, restaurant in enumerate(restaurants, 1):
        name = restaurant.get('name', 'レストラン名不明')
        rating = restaurant.get('rating', '評価なし')
        address = restaurant.get('vicinity', '住所不明')

        print(f"\n【{i}】 {name}")
        print(f"評価: {rating}")
        print(f"住所: {address}")

    print("\n" + "=" * 60)
    print("検索完了")

else:
    print(f"レストラン検索エラー: {places_data.get('status', '不明なエラー')}")
```
